```python
import matplotlib.pyplot as plt

# Data provided
ranks = [1, 3, 5, 10, 20, 30]

# Hit rates for each model
dnn_hits = [0.0000, 0.0397, 0.1587, 0.2778, 0.3175, 0.3968]
snn_hits = [0.0000, 0.0794, 0.1587, 0.1984, 0.4365, 0.7540]
gru_hits = [0.0397, 0.0794, 0.1587, 0.2778, 0.3571, 0.4365]
lstm_hits = [0.0000, 0.0794, 0.1587, 0.1984, 0.4365, 0.7143]

# Creating the plot
plt.figure(figsize=(10, 6))
plt.plot(ranks, dnn_hits, marker='o', label='DNN')
plt.plot(ranks, snn_hits, marker='o', label='SNN')
plt.plot(ranks, gru_hits, marker='o', label='GRU')
plt.plot(ranks, lstm_hits, marker='o', label='LSTM')

# Adding labels and title
plt.xlabel('Rank')
plt.ylabel('Hit Rate')
plt.title('Hit Rate Comparison among Models')
plt.xticks(ranks)
plt.grid(True)

# Adding a legend
plt.legend()
plt.savefig('figure_line.png')

# Show plot
plt.show()
```
